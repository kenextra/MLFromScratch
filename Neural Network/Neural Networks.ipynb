```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.io import loadmat
from utils import predict_nn, display_data
```

# Load Data

```python
data = loadmat("data/ex3data1.mat")
data.keys()
```

```python
X = data['X']
m, n = X.shape
print(f"We have {m} hand written digits examples")
y = data['y']
```

# Load weights of pre-trained model

```python
weights = loadmat("data/ex3weights.mat")
weights.keys()
```

```python
theta1 = weights['Theta1']
theta2 = weights['Theta2']
```

# Prediction with pre-trained model

```python
pred = predict_nn(theta1, theta2, X)
```

```python
print(f'Train Accuracy: {np.mean(pred == y) * 100:.2f}%')
```

# Visualize predictions

```python
rp = np.random.permutation(X.shape[0])
```

```python
for i in range(X.shape[0]):
    print("Displaying Example image")
    
    img = X[rp[i], :].reshape(1, -1)
    
    pred = predict_nn(theta1, theta2, img)
    
    img = np.reshape(img, (20,20), order='F')
    plt.imshow(img, cmap='gray', interpolation = 'bicubic')
    plt.show()
    
    print(f'Neural Network Prediction: {pred} (digit {np.mod(pred, 10)})')
    
    s = input('Paused - press enter to continue, q to exit:')
    if s == 'q':
        break
```
